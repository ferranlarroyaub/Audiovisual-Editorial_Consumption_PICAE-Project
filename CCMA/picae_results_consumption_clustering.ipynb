# 2- Cross validation score i SHAP Value. LightGBM (machine learning)

Utilitzem un algoritme classificador de Machine Learning, anomenat LIGHT GBM, que treballa amb dades de tipus mixt. El que fem s agafar una part de les dades (per exemple, 4/5) i utilitzar-les com a "train". El que fa el model s aprendre i entrenar les dades en base al clster al que pertanyen (per exemple, si aquesta unitat de consum t aquests X atributs, pertany al clster 1, etc). 

Aleshores, l'altre part de les dades (1/5 per ex.) s'utilitza com a "test". El model classificador tracta de predir a quin cluster pertany cadascuna d'aquestes unitats de consum.

El cross-validation score el que fa s mesurar l'encert en la predicci, ja que disposem de les etiquetes reals del clster. Aleshores, es poden comparar les etiquetes reals amb les predites pel model i mesurar la probabilitat d'encert.

El que es fa amb el cross-validation score, s realitzar totes les permutacions possibles amb els conjunts de dades "train i test" i treu el valor mig d'encert en la predicci. 

Un valor prop del 100% significa que els clsters es poden predir amb facilitat i podem assignar unitats de consum noves a clsters amb molt d'encert. Aix vol dir que els clsters sn d'alta qualitat, distingibles i significatius.

## 2.1- Cross validation score

```python
df['player_id']=df['player_id'].astype(str)
df['contingut_id']=df['contingut_id'].astype(str)
df['programa_capitol']=df['programa_capitol'].astype(str)
```

```python
#Setting the objects to category 

lgbm_data = df.copy()
del lgbm_data['Segment']
for c in lgbm_data.select_dtypes(include='object'):
    lgbm_data[c] = lgbm_data[c].astype('category')
```

```python
# Light GBM Classifier.
from lightgbm import LGBMClassifier
from sklearn.model_selection import cross_val_score
import numpy as np
import time

start_lgmb=time.time()

clusters=df['Segment'].tolist()    # Cluster labels (C1, C2...)

clf_kp = LGBMClassifier(colsample_bytree=0.8)   # Use LGBM Classifer to train the data and predict the cluster labels
cv_scores_kp = cross_val_score(clf_kp, lgbm_data, clusters, cv=5, scoring='f1_weighted',n_jobs=-1)  # CV score

print(f'CV F1 score for K-Prototypes clusters is {np.mean(cv_scores_kp)}')

end_lgmb=time.time()
print((end_lgmb-start_lgmb)/60.0, 'minuts')
```

## 2.2- SHAP Value

El shap value s un "tree explainer". Agafa les dades de les unitats de consum i les seves etiquetes dels clsters. Passant-li el model classificador de lightGBM, el que fa es "explicar" l'impacte de cada dimensi. En altres paraules, mesura com de rellevant s cada dimensi a l'hora de predir les etiquetes dels clusters.

Per exemple, si treu la dimensi de l'hora d'inici "min_data", quins resultats obt de la predicci de cada clster? L'arbre va realitzant iteracions anant treien cada dimensi i mesuran l'impacte en cada cluster. D'aquesta manera, un valor alt del SHAP Value d'una dimensi per un determinat clster vol dir que aquella dimensi t un fort impacte a l'hora de predir (i per tant formar i distingir) aquell clster.


```python
import shap

clf_kp.fit(lgbm_data, clusters)      # Fit the data and cluster labels and use the TreeExplainer and obtain SHAP VALUES
explainer_kp = shap.TreeExplainer(clf_kp)
shap_values_kp = explainer_kp.shap_values(lgbm_data)

shap.summary_plot(shap_values_kp, lgbm_data, plot_type="bar", plot_size=(15, 10))
```
